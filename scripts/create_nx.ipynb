# Importing Bib

```python
import numpy as np
import networkx as nx
import igraph as ig
import pickle as pk
from tqdm import tqdm_notebook as tqdm
```

# Defining Constants

```python
IG_NET_NAME = '../data/net/net_ig_cartesian.dat'
NX_NET_NAME = '../data/net/net_nx_cart_cust.gexf'
CONVERT_DICT_NAME = '../data/map/z-x-y_to_m-d-a.dat'
```

# Getting Graph

```python
with open(IG_NET_NAME, 'rb') as f:
    g = pk.load(f, encoding='latin1')

with open(CONVERT_DICT_NAME, 'rb') as f:
    convert = pk.load(f, encoding='latin1')
```

# Converting iGraph to NetworkX

```python
edge_list = [tuple(i) for i in tqdm(g.get_edgelist())]

node_list = [(n, {'c_pos': tuple(i['pos_no']),
                  'n_pos': convert[tuple(i['pos_no'])]}) for n, i in tqdm(enumerate(g.vs))]
```

```python
G = nx.Graph()

G.add_nodes_from(node_list)
G.add_edges_from(edge_list)
```
